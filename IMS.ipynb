```python

USER_CHOICE = """
Enter:
- 'a' to add a new product in inventory
- 'l' to list all products in inventory
- 'e' to edit a record
- 'd' to delete a product from inventory
- 'p' to Purchase
- 'q' to quit

Your choice: """

def menu():
    create_records_file()
    create_sales_file()
    
    user_input = input(USER_CHOICE)
    
    while user_input != 'q':
        if user_input == 'a':
            prompt_add_record()
            
        elif user_input == 'l':
            list_records()

        elif user_input == 'd':
            prompt_delete_record()
            
        elif user_input == 'e':
            prompt_edit_record() 
            
        elif user_input == 'p':
            prompt_purchase_product()
       
        else:
            print("Unknown command, please check and try again")
        
        user_input = input(USER_CHOICE)

features = ["Product Name", "Price", "Stock" , "Description", "Brand"]           
        
def prompt_add_record():
    name = input("Enter Name of product: ")
    brand = input("Enter Name of Brand: ")
    price = input("Enter price of product: ")
    stock = input("Enter stock: ")
    date = input("Enter Expiry Date In (yymmdd): ")
    
    add_to_json(name,brand, price,stock,date)

def list_records():
    for record in get_all_records():
        print(record)

def prompt_delete_record():
    name= input("Enter the product name you want to delete from inventory: ")
    delete_record(name)
    
def prompt_edit_record():
    name = input("Enter the product name you want to edit : ")
    brand = input("Enter the product brand you want to edit : ")
    feature = input("Enter feature you want to edit : ") 
    
    if(feature.title() not in features):
        print("No such feature")
    else:
        new_val = input("Enter new value : ")
        edit_feature(name,brand, feature.title(), new_val)

def prompt_purchase_product():
    name = input("Enter Name of product: ")
    brand = input("Enter Name of Brand: ")
    quantity = int(input("Enter quantity: "))
    
    purchase(name, brand, int(quantity))

import json
import time
records_file = "records.json"
sales_file = "sales.json"

def create_sales_file():                 #to create json file
    with open(sales_file,'w') as file:
        json.dump([],file) 
    
def create_records_file():               #to create json file
    with open(records_file,'w') as file:  
        json.dump([],file) 
    
    
def get_all_records():
    with open(records_file, 'r') as file:
        return json.load(file)

def add_to_json(name ,brand, price, stock, date):
    records = get_all_records()
    records.append({"id":int(time.time()),"Product Name" : name,"Brand" : brand ,"Price" : int(price), "Stock": int(stock), "Expiry Date":date})
    save_records(records)
    
def save_records(records):
    with open(records_file, 'w') as file:
        json.dump(records,file)
    
def delete_record(name):
    records = get_all_records()
    records = [record for record in records if record['Product Name'] != name]
    save_records(records)
      
def edit_feature(name,brand, feature , new_value):
    records = get_all_records()
    
    for record in records:
        if (record['Product Name'] == name and record["Brand"] == brand):
            record[feature] = new_value
    save_records(records)       


def purchase(name , brand , quantity):
    records = get_all_records()

    for record in records:
        if (record['Product Name'] == name and record["Brand"] == brand):
            if (int(record["Stock"]) >= quantity):
                total_price =  record["Price"]*quantity
                print("Your total amount is INR : " , total_price , " Rs")
                
                add_sales_to_json(name ,brand, total_price, quantity)
                
                new_stock = int(record["Stock"]) - quantity
                
                edit_feature(name, brand, "Stock", new_stock)
                
            else:
                print("We dont have that much quantity availabe currently, present stock with us is " , record["Stock"] )
                
    
                
    
def get_all_sales():
    with open(sales_file , 'r') as file :
        return json.load(file)

def add_sales_to_json(name ,brand, price, quantity):
    sales = get_all_sales()
    sales.append({"id":time.ctime(),"Product Name" : name,"Brand" : brand ,"Price" : int(price), "Quantity": int(quantity)})
    save_sales(sales)
    
def save_sales(sales):
    with open(sales_file, 'w') as file:
        json.dump(sales,file)
            
                      
menu()        
```
